## Seaborn Example Gallery

> - https://seaborn.pydata.org/examples/index.html

````python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid")

rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
````

## Matplotlib Example Gallery

> - https://towardsdatascience.com/matplotlib-tutorial-with-code-for-pythons-powerful-data-visualization-tool-8ec458423c5e

````python
import matplotlib.pyplot as plt
x = [1, 2, 3]
x2 = [1, 3, 3.5]
y = [-4, 0, 8]
y2 = [-2, 3, -1]

plt.plot(x, y, label='Line 1')
plt.plot(x2, y2, label='Line 2')

plt.title("Plotting two simple lines")
plt.grid(True)
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.xlim([0, 4])
plt.ylim([-5, 10])
plt.legend()
plt.show()
````

## Plotly Example Gallery

> - https://plotly.com/python/

````python
import plotly.express as px
df = px.data.gapminder()

fig = px.scatter(df.query("year==2007"), x="gdpPercap", y="lifeExp",
	         size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)
fig.show() # no carga la gráfica
````

## Matplotlib

````python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset(name='mpg')

plt.scatter(x='weight',y='acceleration',data=df,c=df['model_year'])

plt.xlabel("weight")
plt.ylabel("acceleration")
plt.legend() # ¿por qué solo aparece acceleration?
plt.show()

````
