# `train_test_split()` & Compare Again

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## `DecisionTreeClassifier()`

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
solut_dtc=model.predict(X_test)

(solut_dtc==y_test).mean()
```

## Automate the Procedure into a function()

```python
def media_aciertos(model):

    model.fit(X_train,y_train)

    solut_dtc=model.predict(X_test)
    
    (solut_dtc==y_test).mean()
    
    resultado = (solut_dtc==y_test).mean()
        
    return resultado
```

## `RandomForestClassifier()`

```python
rdf=RandomForestClassifier()
```

```python
media_aciertos(model = rdf)
```

## `KNeighborsClassifier()`

```python
knb=KNeighborsClassifier()
```

```python
X.columns
```

```python
media_aciertos(model=knb)
```
